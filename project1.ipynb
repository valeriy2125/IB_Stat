```python
import pandas as pd
import os
import glob
```

```python
path = "/home/valeriy/IB/project1/athlete_events"
```

```python
def read(path):
    files = glob.glob(os.path.join(path, "*.csv"))
    pd_function = partial(pd.read_csv, sep = ",")    
    dfs = []
    for f in files:
        df = pd_function(f)
        dfs.append(df)
    athlete_events = pd.concat(dfs)
    return athlete_events
```

```python
events = read(path)
```
